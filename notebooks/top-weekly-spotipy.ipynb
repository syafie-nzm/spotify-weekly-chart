```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from tqdm import tqdm
```

```python
client_id = '2bb47ef8c4c2402cbecb2cb5c48ed809'
client_secret = '<secret>'
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
playlist_link = 'https://open.spotify.com/playlist/37i9dQZEVXbKcS4rq3mEhp?si=1b37d48b74844449'
playlist_URI = playlist_link.split('/')[-1].split('?')[0]
#track_uris = [x["track"]["uri"] for x in sp.playlist_tracks(playlist_URI)["items"]]
```

```python
test = sp.playlist_tracks(playlist_URI)["items"][0]['track']['uri']
audio_features = sp.audio_features(test)
```

```python
track_ls = []
position = 1
for track in tqdm(sp.playlist_tracks(playlist_URI)["items"]):
    #Track URI
    track_uri = track["track"]["uri"]

    #Track name
    track_name = track["track"]["name"]

    #Main Artist
    artist_uri = track["track"]["artists"][0]["uri"]
    artist_info = sp.artist(artist_uri)

    #Name, popularity, genre
    artist_name = track["track"]["artists"][0]["name"]
    artist_pop = artist_info["popularity"]
    artist_genres = artist_info["genres"]

    #Album
    album = track["track"]["album"]["name"]

    #Popularity of the track
    track_pop = track["track"]["popularity"]

    #Track audio features
    audio_features = sp.audio_features(track_uri)
    danceablity = audio_features[0]['danceability']
    energy = audio_features[0]['energy']
    key = audio_features[0]['key']
    loudness = audio_features[0]['loudness']
    mode = audio_features[0]['mode']
    speechiness = audio_features[0]['speechiness']
    acousticness = audio_features[0]['acousticness']
    instrumentalness = audio_features[0]['instrumentalness']
    liveness = audio_features[0]['liveness']
    valence = audio_features[0]['valence']
    tempo = audio_features[0]['tempo']
    duration_ms = audio_features[0]['duration_ms']

    track_ls.append({'Position' : position,
                    'Track URI': track_uri,
                    'Artist': artist_name,
                    'Track Name': track_name,
                    'Album Name': album,
                    'Artist Popularity': artist_pop,
                    'Artist Genre': artist_genres,
                    'Track Popularity': track_pop,
                    'Danceability': danceablity,
                    'Energy': energy,
                    'key': key,
                    'loudness': loudness,
                    'mode': mode,
                    'speechiness': speechiness,
                    'acousticness': acousticness,
                    'instrumentalness': instrumentalness,
                    'liveness': liveness,
                    'valence': valence,
                    'tempo': tempo,
                    'duration_ms': duration_ms
                     })

    position += 1
```

```python
import csv

csv_path = '/content/drive/MyDrive/raw_spotify_2.csv'

with open(csv_path, 'w', newline='') as csvfile:
    fieldnames = track_ls[0].keys()
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for row in track_ls:
        writer.writerow(row)

print(f'Data written to {csv_path}')
```
